## Exercise 16.1 – Image Recognition: Fashion-MNIST Dataset

Keras comes bundled with the Fashion-MNIST dataset of fashion articles which, like the MNIST digits dataset, provides 28×28 grayscale images.

Fashion-MNIST contains clothing article images labeled in 10 categories:

- 0 – T-shirt/top  
- 1 – Trouser  
- 2 – Pullover  
- 3 – Dress  
- 4 – Coat  
- 5 – Sandal  
- 6 – Shirt  
- 7 – Sneaker  
- 8 – Bag  
- 9 – Ankle boot  

The dataset includes:
- 60,000 training images
- 10,000 testing images

In this exercise, the convolutional neural network example is modified to load and process the Fashion-MNIST dataset instead of MNIST. The goal is to compare the model's performance and training time between the two datasets.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import time
```

## Load the Datasets

Both MNIST and Fashion-MNIST contain 28×28 grayscale images.  
They also contain the same number of training and testing samples, which makes them useful for a direct comparison.

```python
# Load MNIST
(X_train_mnist, y_train_mnist), (X_test_mnist, y_test_mnist) = keras.datasets.mnist.load_data()

# Load Fashion-MNIST
(X_train_fashion, y_train_fashion), (X_test_fashion, y_test_fashion) = keras.datasets.fashion_mnist.load_data()
```

## Preprocess the Data

The pixel values range from 0 to 255.  
To help the neural network train more efficiently, the values are normalized to the range 0 to 1.

```python
# Normalize data
X_train_mnist = X_train_mnist.astype("float32") / 255.0
X_test_mnist = X_test_mnist.astype("float32") / 255.0

X_train_fashion = X_train_fashion.astype("float32") / 255.0
X_test_fashion = X_test_fashion.astype("float32") / 255.0
```

## Build the Baseline Convolutional Neural Network

The CNN includes:
- convolution layers to detect image patterns
- max-pooling layers to reduce dimensionality
- dense layers for classification

```python
def create_model():
    model = keras.Sequential([
        layers.Reshape((28, 28, 1), input_shape=(28, 28)),
        layers.Conv2D(32, (3, 3), activation="relu"),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation="relu"),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(128, activation="relu"),
        layers.Dense(10, activation="softmax")
    ])

    model.compile(
        optimizer="adam",
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model
```

## Train and Evaluate the Model on MNIST

```python
mnist_model = create_model()

start_time = time.time()

history_mnist = mnist_model.fit(
    X_train_mnist,
    y_train_mnist,
    epochs=5,
    validation_split=0.1,
    verbose=1
)

mnist_time = time.time() - start_time

mnist_loss, mnist_accuracy = mnist_model.evaluate(X_test_mnist, y_test_mnist, verbose=0)

print("MNIST Accuracy:", mnist_accuracy)
print("MNIST Training Time:", mnist_time)
```

## Train and Evaluate the Model on Fashion-MNIST

```python
fashion_model = create_model()

start_time = time.time()

history_fashion = fashion_model.fit(
    X_train_fashion,
    y_train_fashion,
    epochs=5,
    validation_split=0.1,
    verbose=1
)

fashion_time = time.time() - start_time

fashion_loss, fashion_accuracy = fashion_model.evaluate(X_test_fashion, y_test_fashion, verbose=0)

print("Fashion-MNIST Accuracy:", fashion_accuracy)
print("Fashion-MNIST Training Time:", fashion_time)
```

## Exercise 16.1 Results and Discussion

After running the experiments, the model typically performs better on the MNIST dataset than on the Fashion-MNIST dataset. This is because MNIST digits are simpler and easier to distinguish, while clothing items in Fashion-MNIST often share similar shapes and visual features.

In many cases, MNIST accuracy reaches around 99%, while Fashion-MNIST accuracy is usually lower, often around 90% to 93%, depending on the exact training setup. Training times for the two datasets are generally similar because both contain the same number of images and the same resolution. However, Fashion-MNIST may sometimes take slightly longer to converge because it represents a more difficult classification problem.

```python
print("=== Exercise 16.1 Comparison ===")
print(f"MNIST Accuracy: {mnist_accuracy:.4f}")
print(f"Fashion-MNIST Accuracy: {fashion_accuracy:.4f}")
print(f"MNIST Training Time: {mnist_time:.2f} seconds")
print(f"Fashion-MNIST Training Time: {fashion_time:.2f} seconds")
```

## Exercise 16.4 – Add Another Dense Layer with 4096 Neurons

In this experiment, the neural network is modified by adding an additional Dense layer with **4096 neurons**. This significantly increases the size and complexity of the network.


```python
def create_large_model():
    model = keras.Sequential([
        layers.Reshape((28, 28, 1), input_shape=(28, 28)),
        layers.Conv2D(32, (3, 3), activation="relu"),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation="relu"),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(4096, activation="relu"),
        layers.Dense(128, activation="relu"),
        layers.Dense(10, activation="softmax")
    ])

    model.compile(
        optimizer="adam",
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model
```

## Train the Larger Model on Fashion-MNIST

```python
large_model = create_large_model()

start_time = time.time()

history_large = large_model.fit(
    X_train_fashion,
    y_train_fashion,
    epochs=5,
    validation_split=0.1,
    verbose=1
)

large_time = time.time() - start_time

large_loss, large_accuracy = large_model.evaluate(X_test_fashion, y_test_fashion, verbose=0)

print("Large Model Accuracy:", large_accuracy)
print("Large Model Training Time:", large_time)
```

## Exercise 16.4 Results and Discussion

Adding the extra Dense layer with 4096 neurons greatly increases the number of trainable parameters in the model. Because of this, the larger model usually takes noticeably longer to train than the baseline model.

In some cases, the larger model may slightly improve prediction accuracy because it has more capacity to learn complex features. However, the improvement is often small compared to the increase in training time and computational cost. This means the added complexity does not always produce a meaningful benefit.

```python
print("=== Exercise 16.4 Comparison ===")
print(f"Baseline Fashion-MNIST Accuracy: {fashion_accuracy:.4f}")
print(f"Large Model Accuracy: {large_accuracy:.4f}")
print(f"Baseline Training Time: {fashion_time:.2f} seconds")
print(f"Large Model Training Time: {large_time:.2f} seconds")
```
